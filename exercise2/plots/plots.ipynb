```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = pd.DataFrame({
    "type": ["NN", "NN", "NN", "PSO-NN", "PSO-NN", "PSO-NN"],
    "train_time": [2.2, 0.26, 0.13, 149.39, 79.56, 28.76],
    "acc": [28, 77, 58, 21, 94, 79],
    "dataset": ["Pen Digits", "Breast Cancer", "Glass", "Pen Digits", "Breast Cancer", "Glass"]
})
```

```python
sns.set_theme()

ax = sns.barplot(data, x="dataset", y="train_time", hue="type")
for i in ax.containers:
    ax.bar_label(i)

ax.set(xlabel="Dataset", ylabel="Training Time", title="Train Time per Dataset")
plt.show()
```

```python
ax = sns.barplot(data, x="dataset", y="acc", hue="type")
for i in ax.containers:
    ax.bar_label(i)

ax.set(xlabel="Dataset", ylabel="Accuracy", title="Accuracy per Dataset")
plt.show()
```
